```python
# Importing the neccessary prerequisites for the project
from IPython.display import Markdown, display
MODEL = "llama3.2"
```

```python
# Crafting our prompts for both the system and the user
system_prompt = "You are an email summarizer, your task is to go through a provided email and \
                 summarize it for the end user without missing out on any key details of the email. \
                 Try to make the output as concise as possbile please. Also, after giving a brief summary, \
                 list out the key details of the email with bullet points in case one finds your summary too \
                 chunky or lengthy."

user_prompt = " Subject: Long Time No See… Let’s Fix That! \
                Hey Annie! Wow… it feels like forever since we last caught up! Life’s been a rollercoaster over here. \
                I’ve survived my cooking experiments that may or may not have triggered the smoke alarm, discovered that \
                my dog secretly judges my taste in music, and generally been adulting… kind of. I’ve been thinking it’s \
                high time we catch up properly, share some laughs, and maybe swap embarrassing stories about how we’re still \
                figuring out this whole “growing up” thing. So… I’m inviting you to a little shindig I’m throwing this Friday! \
                It’ll be good vibes only, some food, maybe a little dancing (or just awkwardly flailing—your choice), and \
                definitely a lot of laughing at how ridiculous we both are. Come for the party, stay for the stories, and \
                let’s pretend for a night that we’ve got our lives together. Let me know if you can make it, I promise to \
                save you a snack… maybe even two. Can’t wait to see you and hear everything I’ve been missing out on! \
                Your partner-in-crime, Jessica"
```

```python
# Creating the message list
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
```

```python
# Connecting to the Ollama Server
from openai import OpenAI
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
# chat - we are using a message-based conversation system
# completions - the model generates the next reply in the conversation
# create - runs the request and returns the model's response
# model - selects which AI model will generate the response
# messages - the full conversation input sent to the model, including system instructions (rules/behavior)
# and user input (the actual task/question the model must respond to)
response = ollama.chat.completions.create(model=MODEL, messages=messages)
# response - full model output object
# choices - list of possible generated answers
# [0] - selects the first/best answer
# message - the assistant's reply container
# content - the actual text response from the model
summary = response.choices[0].message.content
display(Markdown(summary))
```
