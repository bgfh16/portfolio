```python
from IPython.display import Markdown, display, update_display
from scraper import fetch_website_contents
from openai import OpenAI
MODEL = "llama3.2"
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
# System Prompt describing the system's role in this generator.
script_system_prompt = """
You are an expert fantasy screenplay writer specializing in Harry Potter-style storytelling.

Your job is to write cinematic, emotionally engaging movie scripts that feel consistent
with the Harry Potter universe, characters, lore, dialogue style, and worldbuilding.

STRICT REQUIREMENTS:
- Write in proper screenplay format
- Use scene headings like: INT. HOGWARTS - NIGHT
- Include dialogue and scene descriptions
- Expand on unresolved mysteries or forgotten storylines
- Keep the tone magical, cinematic, and immersive
- Return ONLY clean Markdown
- Do NOT use code blocks
"""
```

```python
# Scraper function for scraping the main sites containing the most important set of information relating to 
# Harry Potter.
def get_harry_potter_lore():
    main_page = fetch_website_contents(
        "https://en.wikipedia.org/wiki/Harry_Potter"
    )
    hogwarts_page = fetch_website_contents(
        "https://harrypotter.fandom.com/wiki/Hogwarts"
    )
    # Combining all scraped lore into one large context string
    return f"""
    MAIN HARRY POTTER PAGE:
    {main_page}

    HOGWARTS LORE:
    {hogwarts_page}
    """
```

```python
# User prompt that gives the model its actual task using one shot prompting
# and injecting the scraped lore into it.
def get_script_user_prompt():

    lore = get_harry_potter_lore()

    return f"""
    Below is scraped Harry Potter lore and background information.

    Using this information, write a brand new Harry Potter movie screenplay
    based on unresolved mysteries, forgotten magic, or unexplored characters.

    Example screenplay format:

    # Harry Potter and the Cursed Shadows

    ## Opening Scene

    INT. HOGWARTS LIBRARY - NIGHT

    Hermione flips through an ancient dusty book as candlelight flickers.

                        HERMIONE
            This wasn't destroyed...
            Someone hid it.

    A distant whisper echoes through the library halls.

                        RON
            Hermione... did you hear that?

    END EXAMPLE

    Now write a completely original screenplay.

    HARRY POTTER LORE:
    {lore}
    """
```

```python
# Function to generate our custom Harry Potter movie script using a streamed output format
def create_harry_potter_movie():
    stream = ollama.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": script_system_prompt},
            {"role": "user", "content": get_script_user_prompt()}
          ],
        stream=True
    )    
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
create_harry_potter_movie()
```
